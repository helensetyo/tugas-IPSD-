3. cek nilai duplikat dan nilai unik

4. visualisasikan jumlah nilai unik

```python
# Type - Bar Plot
plt.figure(figsize=(15, 4))
sns.countplot(data=data, x='Company', palette='Set2')
plt.title('Count of Company')
plt.xlabel('Company')
plt.ylabel('Count')
plt.show()
```

5. menemukan null values

```python
# Menggunakan Metode 'isnull()' atau 'isna()'
null_values = data.isnull().sum()
print(null_values)
```

6. megethui tipe data

```python
print(data.dtypes)
```

7. melakukan vilter data

```python
# Memfilter data dengan harga di bawah 1000 euros dan RAM 8GB
filtered_price_ram = data[(data['Price_euros'] < 1000) & (data['Ram'] == 8)]
print("\nFiltered Data for Price < 1000 and RAM = 8:")
print(filtered_price_ram)
```

```python
# Memfilter data berdasarkan CPU company Intel
filtered_cpu = data[data['CPU_company'] == 'Intel']
print("\nFiltered Data for Intel CPU:")
print(filtered_cpu)
```

8. membuat box plot

```python
# CPU_freq berdasarkan CPU_company
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, x='CPU_company', y='CPU_freq', palette='magma')
plt.title('Box Plot Frekuensi CPU berdasarkan Perusahaan CPU')
plt.xlabel('Perusahaan CPU')
plt.ylabel('Frekuensi CPU (GHz)')
plt.grid()
plt.show()
```

```python
# RAM berdasarkan OS
plt.figure(figsize=(10, 4))
sns.boxplot(data=data, x='OS', y='Ram', palette='viridis')
plt.title('Box Plot RAM berdasarkan Sistem Operasi')
plt.xlabel('Sistem Operasi')
plt.ylabel('RAM (GB)')
plt.grid()
plt.show()
```

9. korelasi

```python
# Menghitung korelasi antara Price_euros dan Weight
price_weight_corr = data['Price_euros'].corr(data['Weight'])
print(f"Korelasi antara Price_euros dan Weight: {price_weight_corr:.2f}")
```

```python
# Mengatur ukuran figure
plt.figure(figsize=(8, 4))

# Menghitung korelasi dan membuat heatmap
corr_matrix = data.select_dtypes('number').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
